```python
from postprocessing import PostProcess
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('_dm_img.png')
img = img[:,:,::-1]
im = cv2.imread('prediction2.png', 0)

#print(im.shape)

res = PostProcess(im, 40, 16)

plt.figure(figsize=(15,15))
plt.subplot(1,3,1), plt.imshow(img)
plt.subplot(1,3,2), plt.imshow(im, cmap='gray')
plt.subplot(1,3,3), plt.imshow(res, cmap='jet')
plt.show()
```
